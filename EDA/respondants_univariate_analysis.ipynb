```python
def discrete_univariate_analysis(discret_data):
    for col_name in discret_data:
        print("*"*10, col_name ,"*"*10)
        print(discret_data[col_name].agg(['count', 'nunique', 'unique']))
        print("value_counts: \n", discret_data[col_name].value_counts())
        print()
```

```python
discrete_univariate_analysis(data.iloc[ : ,1:9])
```

```python
discrete_univariate_analysis(data.iloc[ : ,9:13])
```

```python
discrete_univariate_analysis(data.iloc[ : ,13:18])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


sns.histplot(data['Gender'], kde=False, bins=5)
plt.xlabel('Gender')
plt.ylabel('Ferquency')
plt.title('Histogram using seaborn')
plt.show()
```

```python
sns.countplot(data['Age'])
plt.xlabel('Frequency')
plt.ylabel('Age')
plt.title('Countlot using seaborn')
plt.show()
```

```python
sns.histplot(data['Employment Status'], kde=False, bins=5)
plt.xlabel('Employment status')
plt.ylabel('Frequency')
plt.title('histplot using seaborn')

plt.show()
```

```python
edu_q=data['Education Qualification '].value_counts()
plt.figure(figsize=(6,4))
plt.pie(edu_q, labels=edu_q.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Pie chart of Educational qualification')
plt.show()
```

```python
sns.barplot(data['Area of Residence'])
plt.xlabel('Frequency')
plt.ylabel('Are')
plt.title('Barplot using seaborn')
plt.show()
```

```python
type_of_vehicles=data.iloc[ : ,7:8].value_counts()
plt.pie(type_of_vehicles, labels=type_of_vehicles.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Pie chart for types of vehicles')
plt.show()
```

```python
sns.countplot(data['Do you have an Electric Vehicle?'])
plt.xlabel('Frequency')
plt.ylabel('Options')
plt.title('Countplot using seaborn')
plt.show()
```
