```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\ANAND\Documents\GitHub\A-study-on-awareness-and-perceptionn-among-the-users-of-EVs\responces.csv")
df = df.drop(columns='Name', axis=1)
df.head()
```

```python
encoder = OneHotEncoder()
df_enc = encoder.fit_transform(df)
print(df_enc)
```

```python
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(df_enc)
X_scaled
```

```python
# Convert sparse matrix to dense NumPy array
X_scaled_dense = X_scaled.toarray()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.decomposition import PCA

# Step 1: Scale the data
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(df_enc)

# Convert sparse matrix to dense
X_scaled_dense = X_scaled.toarray()

# Step 2: Compute the linkage matrix for the dendrogram
Z = linkage(X_scaled_dense, method='ward')

# Step 3: Plot the dendrogram
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
dendrogram(Z)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.show()

# Step 4: Perform hierarchical clustering with a specific number of clusters
n_clusters = 3  # Specify the number of clusters
hc = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
clusters = hc.fit_predict(X_scaled_dense)

# Check cluster labels
print("Cluster labels:", clusters)

# Step 5: Perform PCA for 2D visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled_dense)

# Plot the PCA results with cluster labels
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='Set1', s=100, alpha=0.7)

# Set plot labels and title
plt.title('Hierarchical Clustering (PCA Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
df_clusters = pd.DataFrame(clusters, columns=['Cluster'])

# Count the number of points in each cluster
cluster_counts = df_clusters['Cluster'].value_counts()

# Print the counts
print("Cluster counts:\n", cluster_counts)
```
