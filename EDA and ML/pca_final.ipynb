2. Principal Component Analysis (PCA) Results

Explain the purpose of PCA in your analysis, which is to reduce dimensionality while retaining as much variance as possible.

PCA Analysis:

Applied PCA to reduce the dataset from 36 features to 102 components.

Explained Variance: Describe how much variance each of the components explains. You can include a plot of the explained variance ratio for the first few components to illustrate the contribution of each component.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ANAND\Documents\GitHub\A-study-on-awareness-and-perceptionn-among-the-users-of-EVs\responces.csv")
df = df.drop(columns='Name', axis=1)
df.head()
```

```python
scaler = OneHotEncoder()
df_enc = scaler.fit_transform(df)
print(df_enc)
```

```python
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(df_enc)
X_scaled
```

```python
# Run PCA without specifying the number of components
pca = PCA()
pca.fit(X_scaled)

# Get the explained variance for each component
explained_variance = pca.explained_variance_ratio_
print(explained_variance)

```

```python
# Calculate cumulative variance
cumulative_variance = np.cumsum(explained_variance)
print(cumulative_variance)
```

```python
# Find the number of components to capture 95% of the variance
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components to retain 95% variance: {n_components}")
```

```python
pca1 = PCA(n_components=102)
pca_dt1 = pca1.fit_transform(X_scaled)
```

```python
from sklearn.cluster import KMeans

# Example: Create K-Means clusters
kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
labels = kmeans.fit_predict(X_scaled)  # Fit to the scaled data
```

```python
import matplotlib.pyplot as plt

# Explained variance ratio plot
explained_variance = pca1.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()
```

Key Findings:

The first few principal components account for a significant portion of the total variance. For example, PC1 and PC2 together might explain 50% of the variance, indicating that these components capture essential patterns in the data

```python
import pandas as pd

# Create a DataFrame with cluster assignments
cluster_df = pd.DataFrame(pca_dt1, columns=[f'PC{i+1}' for i in range(pca_dt1.shape[1])])
cluster_df['Cluster'] = labels

# Analyze characteristics of each cluster
cluster_summary = cluster_df.groupby('Cluster').mean()
print(cluster_summary)
```

4. Visualization of Clusters

Provide visual representations of the clusters in relation to the principal components.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a DataFrame for the PCA results
pca_df = pd.DataFrame(pca_dt1[:, :6], columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6'])
pca_df['Labels'] = labels  # Assuming labels is available

# 3D scatter plot for the first three principal components
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot using the first three principal components
scatter = ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], 
                     c=pca_df['Labels'], cmap='viridis', alpha=0.6)

# Set labels and title
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D Scatter Plot of PCA-reduced Data')

# Optional: Add a color bar for the labels
cbar = plt.colorbar(scatter)
cbar.set_label('Cluster Label')

plt.show()
```
