```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\ANAND\Documents\GitHub\A-study-on-awareness-and-perceptionn-among-the-users-of-EVs\responces.csv")
df = df.drop(columns='Name', axis=1)
df.head()
```

```python
scaler = OneHotEncoder()
df_enc = scaler.fit_transform(df)
print(df_enc)
```

```python
scaler = StandardScaler(with_mean=False)
X_scaled = scaler.fit_transform(df_enc)
X_scaled
```

```python
# Run PCA without specifying the number of components
pca = PCA()
pca.fit(X_scaled)

# Get the explained variance for each component
explained_variance = pca.explained_variance_ratio_
print(explained_variance)

```

```python
# Calculate cumulative variance
cumulative_variance = np.cumsum(explained_variance)
print(cumulative_variance)
```

```python
# Find the number of components to capture 95% of the variance
n_components = np.argmax(cumulative_variance >= 0.95) + 1
print(f"Number of components to retain 95% variance: {n_components}")
```

```python
pca1 = PCA(n_components=102)
pca_dt1 = pca1.fit_transform(X_scaled)
```

```python
from sklearn.cluster import KMeans

# Example: Create K-Means clusters
kmeans = KMeans(n_clusters=3)  # Adjust the number of clusters as needed
labels = kmeans.fit_predict(X_scaled)  # Fit to the scaled data
```

```python
import matplotlib.pyplot as plt

# Scatter plot of the first two principal components
plt.figure(figsize=(10, 6))
plt.scatter(pca_dt1[:, 0], pca_dt1[:, 1], c=labels, cmap='viridis', alpha=0.6)

plt.colorbar(label='Cluster Label')  # Optional: Show color mapping to labels
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: First Two Components')
plt.show()
```

```python
import pandas as pd
import seaborn as sns

# Create a DataFrame for the first few components
pca_df = pd.DataFrame(pca_dt1[:, :6], columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6'])
pca_df['Labels'] = labels  # Assuming labels is available

# Pairplot for the first five principal components
sns.pairplot(pca_df, hue='Labels', palette='viridis')
plt.suptitle('Pairwise Scatter Plots of First 5 Principal Components', y=1.02)
plt.show()
```

```python
explained_variance = pca1.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot for the first three components
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca_dt1[:, 0], pca_dt1[:, 1], pca_dt1[:, 2], 
                     c=labels, cmap='viridis', alpha=0.6)

# Add labels and title
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('3D Scatter Plot of First Three Principal Components')

# Optional: Add a color bar for the labels
cbar = plt.colorbar(scatter)
cbar.set_label('Cluster Label')

plt.show()
```

```python
from sklearn.manifold import TSNE

# Apply t-SNE to reduce dimensions to 2
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(pca_dt1)

# Scatter plot of t-SNE results
plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='viridis', alpha=0.6)
plt.colorbar(label='Cluster Label')
plt.title('t-SNE Visualization of PCA-reduced Data')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
