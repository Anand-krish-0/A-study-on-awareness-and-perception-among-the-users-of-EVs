Countings of the ratings of current awreness related questions given by the people. 

```python
#Finding the ratings current awwareness.
sum1a=data['1.1 current awareness'].value_counts()
sum2a=data['1.2 current awareness'].value_counts()
sum3a=data['1.3 current awareness'].value_counts()
sum4a=data['1.4 current awareness'].value_counts()
cur_awr=pd.concat([sum1a, sum2a, sum3a, sum4a], axis=1)
cur_awr=cur_awr.sort_index()
cur_awr.columns.values[0]='aq1'
cur_awr.columns.values[1]='aq2'
cur_awr.columns.values[2]='aq3'
cur_awr.columns.values[3]='aq4'
cur_awr
```

Countings of the ratings of user satisfaction related questions given by the people. 

```python
#Finding the ratings of user satsfaction.
sum1b=data['2.1 user satisfaction'].value_counts()
sum2b=data['2.2 user satisfaction'].value_counts()
sum3b=data['2.3 user satisfaction'].value_counts()
sum4b=data['2.4 user satisfaction'].value_counts()
sum5b=data['2.5 user satisfaction'].value_counts()
use_sat=pd.concat([sum1b, sum2b, sum3b, sum4b, sum5b], axis=1)
use_sat=use_sat.sort_index()
use_sat.columns.values[0]='bq1'
use_sat.columns.values[1]='bq2'
use_sat.columns.values[2]='bq3'
use_sat.columns.values[3]='bq4'
use_sat.columns.values[4]='bq5'
use_sat
```

```python
#Finding the ratings of charging facilities.
sum1c=data['3.1 charging facilities'].value_counts()
sum2c=data['3.2 charging facilities'].value_counts()
sum3c=data['3.3 charging facilities'].value_counts()
sum4c=data['3.4 charging facilities'].value_counts()
cha_fac=pd.concat([sum1c, sum2c, sum3c, sum4c], axis=1)
cha_fac=cha_fac.sort_index()
cha_fac.columns.values[0]='cq1'
cha_fac.columns.values[1]='cq2'
cha_fac.columns.values[2]='cq3'
cha_fac.columns.values[3]='cq4'
cha_fac
```

Countings of the ratings of cost consideration related questions given by the people. 

```python
#Finding the ratings of cost consideration.
sum1d=data['4.1 cost consideration'].value_counts()
sum2d=data['4.2 cost consideration'].value_counts()
sum3d=data['4.3 cost consideration'].value_counts()
sum4d=data['4.4 cost consideration'].value_counts()
sum5d=data['4.5 cost consideration'].value_counts()
cos_con=pd.concat([sum1d, sum2d, sum3d, sum4d, sum5d], axis=1)
cos_con=cos_con.sort_index()
cos_con.columns.values[0]='dq1'
cos_con.columns.values[1]='dq2'
cos_con.columns.values[2]='dq3'
cos_con.columns.values[3]='dq4'
cos_con.columns.values[4]='dq5'
cos_con
```

Countings of the ratings of cost consumer preference related questions given by the people. 

```python
#Finding the ratings of consumer priference .
sum1e=data['5.1 consumer preference'].value_counts()
sum2e=data['5.2 consumer preference'].value_counts()
sum3e=data['5.3 consumer preference'].value_counts()
sum4e=data['5.4 consumer preference'].value_counts()
sum5e=data['5.5 consumer preference'].value_counts()
con_pre=pd.concat([sum1e, sum2e, sum3e, sum4e, sum5e], axis=1)
con_pre=con_pre.sort_index()
con_pre.columns.values[0]='eq1'
con_pre.columns.values[1]='eq2'
con_pre.columns.values[2]='eq3'
con_pre.columns.values[3]='eq4'
con_pre.columns.values[4]='eq5'
con_pre
```

Countings of the ratings of environmental consideration related questions given by the people. 

```python
#Finding the ratings of environmental consideratioin.
sum1f=data['6.1 evm consideration'].value_counts()
sum2f=data['6.2 evm consideration'].value_counts()
sum3f=data['6.3 evm consideration'].value_counts()
sum4f=data['6.4 evm consideration'].value_counts()
evm_con=pd.concat([sum1f, sum2f, sum3f, sum4f], axis=1)
evm_con=evm_con.sort_index()
evm_con.columns.values[0]='fq1'
evm_con.columns.values[1]='fq2'
evm_con.columns.values[2]='fq3'
evm_con.columns.values[3]='fq4'
evm_con
```

Merging the frequency table of the seperated topics in to a single table.

```python
df=pd.concat([cur_awr,use_sat,cha_fac,cos_con,con_pre,evm_con], axis=1, ignore_index=False)
df
```

```python
df['ratings']=['1','2','3','4','5']
df.set_index('ratings', inplace=True)
df
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(df)
```
